# Decision tree algorithm

```python
from sklearn.tree import DecisionTreeClassifier
de= DecisionTreeClassifier(class_weight='balanced')
de.fit(train_kn_x, train_kn_y)
pred_test=de.predict(test_kn_x)
from sklearn.metrics import confusion_matrix
tab1 = confusion_matrix(test_kn_y,pred_test)
print(tab1)
print(classification_report(test_kn_y, pred_test))
print('Accuracy is --->', tab1.diagonal().sum()*100/tab1.sum())
```

# Ada boost on decision tree

```python
from sklearn.ensemble import AdaBoostClassifier
abc = AdaBoostClassifier(de)
abc.fit(train_kn_x, train_kn_y)
pred_test=abc.predict(test_kn_x)
from sklearn.metrics import confusion_matrix
tab1 = confusion_matrix(test_kn_y,pred_test)
print(tab1)
print(classification_report(test_kn_y, pred_test))
print('Accuracy is --->', tab1.diagonal().sum()*100/tab1.sum())
```
